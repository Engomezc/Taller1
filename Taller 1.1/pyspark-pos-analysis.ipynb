```python
import nltk
from pyspark import SparkContext
from nltk import pos_tag, word_tokenize
```

```python
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download('gutenberg')
```

```python
# Entry point for working with RDD
sc = SparkContext(appName = "pyspark-pos-analysis")
```

```python
# Loading a text file
rdd_reuters = sc.textFile("./data/reuters.txt")
```

```python
rdd_reuters.count()
```

```python
text = nltk.corpus.gutenberg.words("reuters.txt")
nltk.pos_tag(text)
```

```python
text = nltk.corpus.gutenberg.words("prueba.txt")
nltk.pos_tag(text)
```
